```python
import numpy as np
import matplotlib.pyplot as plt
import math
import astropy.units as u
from astropy.constants import c,k_B,h
```

```python
farray = np.arange(0.001E14,15.0E14,0.1E14)*u.Hz
T = 5800*u.K
x = (h*farray)/(k_B*T)
```

$$
B(\nu,T) = \frac{2 h \nu^3}{c^2 (e^{\frac{h \nu}{k T}} -1)}
$$

```python
energydensity = (2*h*farray**3)/(c**2*(np.exp(x)-1))
```

```python
fig,ax = plt.subplots()
ax.plot(x,energydensity)
ax.set(xlabel='hv/kT',ylabel='Spectral Radiance per unit freq. J/m^2',title='Blackbody Radiation T=5800')
plt.show()
```
